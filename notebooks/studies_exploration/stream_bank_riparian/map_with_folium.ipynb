```python
import folium
import pandas as pd
```

```python
input_tsv = "stream_bank_riparian.tsv"
lat_col = "lat"
lon_col = "lon"
id_col = "sample name"
```

```python
df = pd.read_csv(input_tsv, sep="\t")
```

```python
# Extract "stream" or "sediment" from the sample name using regular expressions
df['category'] = df['sample name'].str.extract(r'-(Stream|Riparian) Sediment$')[0].replace({'Stream': 'stream', 'Riparian': 'riparian'})
```

- 'OpenStreetMap' (default)
- 'Mapbox Bright'
- 'Esri WorldImagery'

```python
# Create a folium map centered around the average latitude and longitude using OpenStreetMap
map_center = [df['lat'].mean(), df['lon'].mean()]
mymap = folium.Map(location=map_center, zoom_start=10, tiles='OpenStreetMap')
```

```python
# Define a color mapping for the categories (Stream and Sediment)
category_colors = {'stream': 'blue', 'riparian': 'green'}
```

```python

# Add markers with color based on the category (Stream or Sediment)
for _, row in df.iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']],
        icon=folium.Icon(color=category_colors.get(row['category'], 'gray'))  # Default to gray if category is not found
    ).add_to(mymap)
```
