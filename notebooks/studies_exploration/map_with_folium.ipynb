```python
import folium
import pandas as pd
```

```python
input_tsv = "stream_bank_riparian/stream_bank_riparian.tsv"
output_html="stream_bank_riparian/stream_bank_riparian.html"

# input_tsv = "emp_500_ng/MacRae-Crerar/macrae_crerar_attributes.tsv"
# output_html="emp_500_ng/MacRae-Crerar/macrae_crerar.html"

# input_tsv = "emp_500_ng/myrold/myrold_attributes_with_land_cover.tsv"
# output_html="emp_500_ng/myrold/myrold_attributes.html"
```

```python
lat_col = "lat"
lon_col = "lon"
id_col = "sample name"
category_col = "category"

# lat_col = "latitude_decimal"
# lon_col = "longitude_decimal"
# id_col = "accession"
# category_col = "cluster"

# lat_col = "latitude"
# lon_col = "longitude"
# id_col = "accession"
# category_col = "envo_label"
```

```python
df = pd.read_csv(input_tsv, sep="\t")
```

```python
# Get unique categories from the category column
categories = df[category_col].unique()
```

```python
# Define a basic list of colors (extend it if necessary)
color_list = ['red', 'green', 'blue', 'orange', 'purple', 'white', 'black', 'beige', 'gray']

# Map each category to a color by cycling through the color_list
category_colors = {category: color_list[i % len(color_list)] for i, category in enumerate(categories)}
```

- 'OpenStreetMap' (default)
- 'Mapbox Bright'
- 'Esri WorldImagery'

```python
# Create a folium map centered around the average latitude and longitude using OpenStreetMap
map_center = [df[lat_col].mean(), df[lon_col].mean()]
mymap = folium.Map(location=map_center, zoom_start=10, tiles='OpenStreetMap')
```

```python
# Add markers with color based on the category
for _, row in df.iterrows():
    folium.Marker(
        location=[row[lat_col], row[lon_col]],
        popup=row[id_col],  # Show sample name in popup
        icon=folium.Icon(color=category_colors.get(row[category_col], 'gray'))  # Default to gray if category is not found
    ).add_to(mymap)
```
