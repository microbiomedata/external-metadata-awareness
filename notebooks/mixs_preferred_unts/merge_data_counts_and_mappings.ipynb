```python
import pandas as pd
```

```python
nmdc_mongo_slot_usage = pd.read_csv("nmdc.flattened_biosample_field_counts.csv", sep=",")
```

```python
nmdc_mongo_slot_usage.columns = ['path', 'count']
```

```python
nmdc_mongo_slot_usage['slot'] = nmdc_mongo_slot_usage['path'].str.replace(r'(_has|_term|_latitude|_longitude).*$', '',
                                                                          regex=True)
```

```python
nmdc_mongo_slot_usage = nmdc_mongo_slot_usage[['slot', 'count']]
```

```python
nmdc_mongo_slot_usage = nmdc_mongo_slot_usage.copy()
nmdc_mongo_slot_usage.drop_duplicates(inplace=True)
```

```python
nmdc_mongo_slot_usage = nmdc_mongo_slot_usage[nmdc_mongo_slot_usage['slot'] != '_id']
```

```python
nmdc_mongo_slot_usage = nmdc_mongo_slot_usage.loc[nmdc_mongo_slot_usage.groupby("slot")["count"].idxmax()].reset_index(
    drop=True)
```

```python
nmdc_mongo_slot_usage = nmdc_mongo_slot_usage.sort_values(by=['count', 'slot'], ascending=[False, True])
```

```python
nmdc_mongo_slot_usage
```

```python
nmdc_submission_slot_usage = pd.read_csv("misc_metadata.submission_biosample_slot_counts.csv", sep=",")
```

```python
nmdc_submission_slot_usage.drop(columns=["_id"], inplace=True)
```

```python
nmdc_submission_slot_usage = nmdc_submission_slot_usage[['field', 'count']]
```

```python
nmdc_submission_slot_usage.columns = ['slot', 'count']
```

```python
nmdc_submission_slot_usage = nmdc_submission_slot_usage.sort_values(by=['count', 'slot'], ascending=[False, True])
```

```python
nmdc_submission_slot_usage
```

```python
exploded_mixs_preferred_units = pd.read_csv("exploded_mixs_preferred_units.tsv", sep="\t")
```

```python
exploded_mixs_preferred_units
```

```python
nmdc_mongo_slots_mixs_prefunits_ucum = nmdc_mongo_slot_usage.merge(exploded_mixs_preferred_units, how='outer')
```

```python
nmdc_mongo_slots_mixs_prefunits_ucum
```

```python
nmdc_mongo_slots_mixs_prefunits_ucum.to_csv("nmdc_mongo_slots_mixs_prefunits_ucum.tsv", sep="\t", index=False)
```

```python
nmdc_submission_slots_mixs_prefunits_ucum = nmdc_submission_slot_usage.merge(exploded_mixs_preferred_units, how='outer')
```

```python
nmdc_submission_slots_mixs_prefunits_ucum.to_csv("nmdc_submission_slots_mixs_prefunits_ucum.tsv", sep="\t", index=False)
```
